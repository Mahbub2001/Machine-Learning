```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Boston_Housing.csv")
dataset=pd.DataFrame(df)
dataset
```

```python
dataset = df.drop(columns=['Unnamed: 14'])
```

```python
dataset.head()
```

```python
#independent features and dependent features
X=dataset.iloc[:,0:13]
Y=dataset.iloc[:,-1]
```

```python
Y
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state=42)
```

```python
X_train
```

```python
##standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
# scaler.inverse_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,Y_train)
```

```python
mse = cross_val_score(regression, X_train, Y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
np.mean(mse) #should be less as possible
```

```python
reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

```python
sns.distplot(reg_predict-Y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test, reg_predict)
```
